### Calculating  the distorsion function

```python
# Distorsion function
from scipy.spatial.distance import cdist

def J (X, C):
    # cdist (X, C) --> calcule toutes les distances entre les points X et les divers Centroid
    # np.min --> prend le minimum de chacune des distances, c'est la définition d'appartenance à une classe
    # sum --> additionne les distances minimale
    #
    # il faudrait diviser par la taille du dataset mais comme c'est une constante, pas nécessaire en pratique
    d = sum(np.min(cdist(X, C, 'euclidean'), axis=1))
    return d
```

```python
from sklearn.cluster import KMeans
# from sklearn import metrics

distortions = []
cluster_range = range(1,10)
for i in cluster_range:
    print("with", i, "centroid(s)")
    # Initializing KMeans and fit with inputs
    kmeans_Models = KMeans(n_clusters=i)
    kmeans = kmeans_Models.fit(Features_pca)
    # Predicting the clusters
    labels = kmeans.predict(Features_pca)
    # Getting the cluster centers
    C = kmeans.cluster_centers_
    # Plot the result
    # original
    fig = plt.figure(figsize = (20, 5))
    ax = fig.add_subplot(121)
    ax.scatter(Features_pca[:, 0], Features_pca[:, 1], s=5, c=labels)
    ax.scatter(C[:, 0], C[:, 1], marker='*', c=["#0000FF"], s=5000,alpha=0.2)
    
    distortions.append(J(Features_pca, C))

    fig.add_subplot(122, XLim=[0, 10])
    plt.plot(range(1,i+1), distortions, 'bx-')
    plt.show()
```
