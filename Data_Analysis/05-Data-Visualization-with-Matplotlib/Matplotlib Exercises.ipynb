# Matplotlib Exercises

#### Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Import matplotlib.pyplot as plt and set %matplotlib inline

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show() for non-notebook users
```

## Exercise 1

I followe along with these steps: 
*  Create a figure object called fig using plt.figure() 
*  Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. 
*  Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Exercise 2

## Exercise 3

I Cceate the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

 Now I use x,y, and z arrays to recreate the plot below.

```python
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

## Exercise 4

I use plt.subplots(nrows=1, ncols=2) to create the plot below:

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

Now I plot (x,y) and (x,z) on the axes. Play around with the linewidth and style.

```python
axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[1].plot(x,z,color="red", lw=3, ls='-')
fig
```

I can resize the plot by adding the figsize() argument in plt.subplots().

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
